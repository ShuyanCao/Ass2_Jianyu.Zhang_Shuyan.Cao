#3.	Methodology

##1) Content-based Algorithm

###a)	Algorithms implementation 
* In this part, the main algorithms implemented here is the cosine similarity matrix. Cosine similarity is a common approach to determine the similarity of the documents based on counting the maximum number of same words in both documents. In terms of mathematics, cosine similarity calculates the cosine of the angle between two vectors in a multidimensional space. In this project, the two vectors are the arrays containing words according to the features defined in the dataset. When we plot multidimensional space, the cosine similarity captures the documents’ angles. 

###b) Build and Train Models
* To build the model, we will use the CountVectorizer from scikit-learn package. Within this function, the keywords from the datasets related to features of movie will be selected and the frequency of each word will be counted in the matrix. After this process is completed, we do the main step which is creating matrix by applying the cosine similarity.  Then, list of indices will be created for the match of movie recommendations. Finally, after all the preparation work is completed, it will be the last step to build the real recommendation function for movies. In the recommendation function, the matching movie will be retrieved firstly. Then, cosine similarity scores will be in descending order. The last step in the function is that the display of the recommendation result of most similar movies.
